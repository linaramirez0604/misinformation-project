
### 0.3. Variables for Regression Model 

```python
# Controls
#df2[(np.abs(stats.zscore(df2['age'])) < 3)]
C = df2[['age', 'female', 'high_school', 'postgraduate', 'technical_educ', 'undergraduate', 'bog_dept']]
```

**Regression Model 1** 

$ Y = \alpha + \beta D + \gamma C + \mu $

$Y =$ outcome

$D = 1$ if any treatment. 

$C =$ Controls: Age, Gender, Education level and Department (Bog vs. the rest)

```python
#Regression Model 1 Code
D = df2[['any_treatment']]
#X = D + C
X = D.join(C)
X = sm.add_constant(X)
```

**Regression Model 2** 

$ Y = \alpha + \beta_1 D_1 + \beta_2 D_2 + \beta_3 D_1 \times D_2 + \gamma C + \mu $

$Y =$ outcome

$D_1 = 1$ if any video.

$D_2 = 1$ if any form. 

$C =$ Controls: Age, Gender, Education level and Department (Bog vs. the rest)

```python
#Regression Model 2 Code
D1 = df2[['any_video']]
D2 = df2[['any_form']]
D3 = df2[['any_video_form']]
D = D1.join(D2)
D = D.join(D3)
X = D.join(C)
X = sm.add_constant(X)
```

**Regression Model 3** 

$ Y = \alpha + \beta_1 D_1 + \beta_2 D_2 + ... + \beta_12 D_{12} + \gamma C + \mu $

$Y =$ outcome

$D_1 = 1$ if only form ambiguity 

$D_2 = 1$ if only form dehumanization

...

$D_{12} = 1$ if video+form = trust 


$C =$ Controls: Age, Gender, Education level and Department (Bog vs. the rest)

```python
#Regression Model 3 Code
D = df2[['treatment_only_f_ambiguity', 'treatment_only_f_dehumanization',
       'treatment_only_f_discrimination', 'treatment_only_f_trust',
       'treatment_only_v_ambiguity', 'treatment_only_v_dehumanization',
       'treatment_only_v_discrimination', 'treatment_only_v_trust',
       'treatment_v+f_dehumanization', 'treatment_v+f_ambiguity',
       'treatment_v+f_discrimination', 'treatment_v+f_trust']] 
X = D.join(C)
X = sm.add_constant(X)
```

**Regression Model 1** 

$ Y = \alpha + \beta D + \gamma C + \mu $

$Y =$ all fake reliable or unreliable; all true reliable or unreliable. (D=1 if 4 or 5 0 otherwise ;D=1 if 1 or 2 0 otherwise) 

$D = 1$ if any treatment. 

$C =$ Controls: Age, Gender, Education level and Department (Bog vs. the rest)

```python
#Declaring the variables of the model

Y1 = df2['d_all_fake_reliable']
Y2 = df2['d_all_true_reliable']
Y3 = df2['d_all_fake_unreliable']
Y4 = df2['d_all_true_unreliable']

D = df2[['any_treatment']]
#X = D + C
X = D.join(C)
X = sm.add_constant(X)
```

```python
# Fake news - reliable measure (Y=1 if response=4 or 5) 
est = sm.OLS(Y1, X)
est1 = est.fit()
#est1.summary()

coef_1 = pd.DataFrame(est1.summary().tables[1].data)

# Column name 
coef_1.columns = coef_1.iloc[0]

# Drop the extra row with column labels
coef_1=coef_1.drop(0)

# Keep only the any_treatment 
coef_1.drop(coef_1.index[[0, 2, 3, 4, 5, 6, 7, 8]], inplace = True)

# Set index to variable names 
coef_1 = coef_1.set_index(coef_1.columns[0])

# Change datatype from object to float
coef_1 = coef_1.astype(float)

coef_1.rename(index={'any_treatment':'Reliable (fake)'},inplace=True)
```

```python
# True news - reliable measure (Y=1 if response=4 or 5) . 
est = sm.OLS(Y2, X)
est2 = est.fit()
#est2.summary()

coef_2 = pd.DataFrame(est2.summary().tables[1].data)

# Column name 
coef_2.columns = coef_2.iloc[0]

# Drop the extra row with column labels
coef_2=coef_2.drop(0)

# Keep only the any_treatment 
coef_2.drop(coef_2.index[[0, 2, 3, 4, 5, 6, 7, 8]], inplace = True)

# Set index to variable names 
coef_2 = coef_2.set_index(coef_2.columns[0])

# Change datatype from object to float
coef_2 = coef_2.astype(float)

coef_2.rename(index={'any_treatment':'Reliable (true)'},inplace=True)
```

```python
#Function for the plot
# Source: https://medium.com/analytics-vidhya/create-your-own-coefficient-plot-function-in-python-aadb9fe27a77
def coefplot(coef_df):

    # Get errors; (coef - lower bound of conf interval)
    errors = coef_df['coef'] - coef_df['[0.025']
    
    # Append errors column to dataframe
    coef_df['errors'] = errors

    # Sort values by coef ascending
    #coef_df = coef_df.sort_values(by=['coef'])
    coef_df = coef_df.sort_index()


    ### Plot Coefficients ###

    # x-labels
    variables = list(coef_df.index.values)
    
    # Add variables column to dataframe
    coef_df['variables'] = variables
    
    # Set sns plot style back to 'poster'
    # This will make bars wide on plot
    sns.set_context("poster")

    # Define figure, axes, and plot
    fig, ax = plt.subplots(figsize=(15, 10))
    
    # Error bars for 95% confidence interval
    # Can increase capsize to add whiskers
    coef_df.plot(x='variables', y='coef', kind='bar',
                 ax=ax, color='none', fontsize=22, 
                 ecolor='lightskyblue',capsize=0,
                 yerr='errors', legend=False)
    
    # Set title & labels
    #plt.title('Fake News - Any treatment',fontsize=26)
    ax.set_ylabel('Coefficients with 95% CI',fontsize=22)
    ax.set_xlabel('',fontsize=22)
    
    # Coefficients
    ax.scatter(x=pd.np.arange(coef_df.shape[0]), 
               marker='o', s=80, 
               y=coef_df['coef'], color='lightskyblue')
    
    # Line to define zero on the y-axis
    ax.axhline(y=0, linestyle='--', color='slateblue', linewidth=1)
    
    return plt.show()
```
